## 將全部訓練影像的路徑生成train.txt檔給yolo訓練

```python
import os
outpath = './cfg/train.txt'
f = open(outpath, 'w')
for root, dirs, files in os.walk(os.getcwd()+'/NEU-DET/train/images/patches/'):    #相對路徑也可以
    for filename in files:
        if filename.endswith('jpg'):
            filepath = os.path.join(root, filename)
            print(filepath)
            f.write(filepath+'\n')
f.close()
```
